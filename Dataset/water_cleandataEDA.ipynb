```python
# merge all of the EDA data (CSVs) into a single CSV file 

import os
import pandas as pd
import glob
import xlsxwriter # this needed pip install first. Only pip install in myenv. 

path = 'waterEDAresult' # all of the exported CSVs above are grouped into a single folder named 'waterEDAresult'
all_files = glob.glob(os.path.join(path, "*.csv"))

writer = pd.ExcelWriter('EDAresult.xlsx', engine='xlsxwriter')

for f in all_files:
    df = pd.read_csv(f)
    df.to_excel(writer, sheet_name=os.path.splitext(os.path.basename(f))[0], index=False)

writer.save()

# source:https://stackoverflow.com/questions/51964001/merging-multiple-csv-files-into-separate-tabs-of-a-spreadsheet-in-python
```
