```python
import pandas as pd
```

```python
data = pd.read_csv('Water_Prevalence_10__commat_2021-09-29.csv')

data
```

```python
index_alpha = pd.read_csv('indices/Water_Prevalence_10__commat_2021-09-29_alpha.csv')

index_alpha = index_alpha.drop(columns=['Sample_ID']) 

index_alpha
```

```python
index_jaccard = pd.read_csv('indices/Water_Prevalence_10__commat_2021-09-29_Jaccard.csv')

# drop all except the first column
index_jaccard = index_jaccard.loc[:, index_jaccard.columns.intersection(['SW_S19S.0003_U_1'])] 

# rename column name
index_jaccard.rename(columns = {'SW_S19S.0003_U_1':'Jaccard'}, inplace = True)

index_jaccard
# https://stackoverflow.com/questions/45846189/how-to-delete-all-columns-in-dataframe-except-certain-ones
```

```python
index_ochiai = pd.read_csv('indices/Water_Prevalence_10__commat_2021-09-29_Ochiai.csv')

# drop all except the first column
index_ochiai = index_ochiai.loc[:, index_ochiai.columns.intersection(['SW_S19S.0003_U_1'])] 

# rename column name
index_ochiai.rename(columns = {'SW_S19S.0003_U_1':'Ochiai'}, inplace = True)

index_ochiai
```

```python
index_sorensen = pd.read_csv('indices/Water_Prevalence_10__commat_2021-09-29_Sorensen.csv')

# drop all except the first column
index_sorensen  = index_sorensen .loc[:, index_sorensen .columns.intersection(['SW_S19S.0003_U_1'])] 

# rename column name
index_sorensen.rename(columns = {'SW_S19S.0003_U_1':'Sorensen'}, inplace = True)

index_sorensen
```

```python
# merging two csv files
# https://stackoverflow.com/questions/23891575/how-to-merge-two-dataframes-side-by-side

result = pd.concat([data, index_sorensen, index_alpha, index_jaccard, index_ochiai], axis=1, ignore_index=False)

result
```

```python
# export the merged csv file 

result.to_csv('water_merged_indices.csv', index=False)
```
