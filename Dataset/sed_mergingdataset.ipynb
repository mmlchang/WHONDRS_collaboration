```python
import pandas as pd
```

```python
data = pd.read_csv('Sediment_Prevalence_10__commat_2021-09-29.csv')

data
```

```python
index_alpha = pd.read_csv('indices/Sediment_Prevalence_10__commat_2021-09-29_alphacsv.csv')

index_alpha = index_alpha.drop(columns=['Sample_ID']) 

index_alpha
```

```python
index_jaccard = pd.read_csv('indices/Sediment_Prevalence_10__commat_2021-09-29_Jaccard.csv')

# drop all except the first index column
index_jaccard = index_jaccard.loc[:, index_jaccard.columns.intersection(['SED_S19S.0001_D_Field'])] 

# rename column name
index_jaccard.rename(columns = {'SED_S19S.0001_D_Field':'Jaccard'}, inplace = True)

index_jaccard
# https://stackoverflow.com/questions/45846189/how-to-delete-all-columns-in-dataframe-except-certain-ones
```

```python
index_ochiai = pd.read_csv('indices/Sediment_Prevalence_10__commat_2021-09-29_Ochiai.csv')

# drop all except the first column
index_ochiai = index_ochiai.loc[:, index_ochiai.columns.intersection(['SED_S19S.0001_D_Field'])] 

# rename column name
index_ochiai.rename(columns = {'SED_S19S.0001_D_Field':'Ochiai'}, inplace = True)

index_ochiai
```

```python
index_sorensen = pd.read_csv('indices/Sediment_Prevalence_10__commat_2021-09-29_Sorensen.csv')

# drop all except the first column
index_sorensen  = index_sorensen .loc[:, index_sorensen .columns.intersection(['SED_S19S.0001_D_Field'])] 

# rename column name
index_sorensen.rename(columns = {'SED_S19S.0001_D_Field':'Sorensen'}, inplace = True)

index_sorensen
```

```python
# merging two csv files
# https://stackoverflow.com/questions/23891575/how-to-merge-two-dataframes-side-by-side

result = pd.concat([data, index_sorensen, index_alpha, index_jaccard, index_ochiai], axis=1, ignore_index=False)

result
```

```python
# export the merged csv file 

result.to_csv('sediment_merged_indices.csv', index=False)
```

```python
merged = pd.read_csv('sediment_merged_indices.csv')

merged
```

```python
# get the metadata 

metadata = pd.read_csv('metdom_sed.csv')

# drop other columns except the metadata 
metadata  = metadata.loc[:, metadata.columns.intersection(['sample.type',	'location.id',	'sampling.type',	
                                                           'Stream_Order',	'Country',	'General_Vegetation',	
                                                           'Hydrogeomorphology',	'River_Gradient',	'Intermittent_or_Perennial',	
                                                           'Sediment_Bedrock',	'Distance_DS.and.MS_meters',	'Distance_MS.and.US_meters',	
                                                           'Primary.Sources.Flow.Variation',	'pH',	'Temp_degC',	
                                                           'DO_perc.sat',	'DO_mg.per.L',	'Algal.Mat.Coverage',	'Depositional.Type',	
                                                           'Macrophyte.Coverage',	'Sunlight.Access_Perc.Canopy.Cover',	'Water.Column.Height_cm',	
                                                           'Approx.Distance.From.Gauge_meters',	'US.or.DS.of.Gauge',	'Precipitation.Within.Week',	
                                                           'Number.Days.Since.Precip',	'Dam.Upstream.Site',	'Contamination.Source.Upstream',	
                                                           'Contamination.Source',	'resprate_mg.L.h',	'resp.inc.time.min', 'NPOC_mg.L.asC	del2H_permil', 
                                                           'del18O_permil',	'Cl_mgL',	'SO4_mgL',	'NO3_mgL',	'NO2_mgL',	'F_mgL'])]  

metadata
```

```python
# concat metadata with molecular binary and indices
result_met = pd.concat([result, metadata], axis=1, ignore_index=False)

result_met
```

```python
# export the merged csv file - metadata added

result_met.to_csv('sediment_merged_metadataindices.csv', index=False)
```
